### 9.1 - Archive Exploration

This dataset explores populations of Snowshoe hares, a keystone prey species, in the Bonanza Creek Experimental in Alaska. The data were collected from 1999-2012 through capture-recapture of Snowshoe hares at 5 locations in Tanana Valley to look at population densities. The archive doesn't state whether or not the data is sensitive. There is one publication associated with the data, titled "Comparison of snowshoe hare populations in Interior Alaska", which used the data as part of a master's thesis at University of Alaska, Fairbanks. 

- archive : https://portal.edirepository.org/nis/mapbrowse?packageid=knb-lter-bnz.55.22
- citation: Kielland, K., F.S. Chapin, R.W. Ruess, and Bonanza Creek LTER. 2017. Snowshoe hare physical data in Bonanza Creek Experimental Forest: 1999-Present ver 22. Environmental Data Initiative. https://doi.org/10.6073/pasta/03dce4856d79b91557d8e6ce2cbcdc14 (Accessed 2023-10-19).
- date of access: 10/19/2023
